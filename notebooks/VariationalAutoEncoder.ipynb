```python
import numpy as np
import six

import chainer
from chainer import computational_graph as c
from chainer import cuda
import chainer.functions as F
from chainer import optimizers

import data

import matplotlib.pyplot as plt
from matplotlib import cm
%matplotlib inline

from climin.util import iter_minibatches
from breze.learn.utils import tile_raster_images
```

```python
mnist = data.load_mnist_data()
mnist['data'] = mnist['data'].astype(np.float32)
mnist['data'] /= 255
mnist['target'] = mnist['target'].astype(np.int32)

N = 60000
x_train, x_test = np.split(mnist['data'],   [N])
y_train, y_test = np.split(mnist['target'], [N])
N_test = y_test.size
```

```python
batch_size = 250
n_epochs = 100

n_hiddens_recog = [512]
n_hiddens_gen = [512]
n_latents = 64

n_layers_recog = len(n_hiddens_recog)
n_layers_gen = len(n_hiddens_gen)
```

```python
layers = {}
```

```python
# Recognition model.

rec_layer_sizes = [(x_train.shape[1], n_hiddens_recog[0])]
rec_layer_sizes += zip(n_hiddens_recog[:-1], n_hiddens_recog[1:])
rec_layer_sizes += [(n_hiddens_recog[-1], n_latents * 2)]

for i, (n_incoming, n_outgoing) in enumerate(rec_layer_sizes):
    layers['recog_%i' % i] = F.Linear(n_incoming, n_outgoing)
```

```python
# Generating model.

gen_layer_sizes = [(n_latents, n_hiddens_gen[0])]
gen_layer_sizes += zip(n_hiddens_gen[:-1], n_hiddens_gen[1:])
gen_layer_sizes += [(n_hiddens_gen[-1], x_train.shape[1])]

for i, (n_incoming, n_outgoing) in enumerate(gen_layer_sizes):
    layers['gen_%i' % i] = F.Linear(n_incoming, n_outgoing)
```

```python
print rec_layer_sizes
print gen_layer_sizes
```

```python
model = chainer.FunctionSet(**layers)
```

```python
class StandardNormalSamples(chainer.function.Function):

    def forward_cpu(self, inputs):
        return np.random.standard_normal(inputs[0].shape).astype('float32'),
```

```python
def sample(latent_data, train=True):
    inpt = chainer.Variable(latent_data)
    chain = [inpt]        
    for i in range(n_layers_gen):
        chain.append(F.relu(getattr(model, 'gen_%i' % i)(chain[-1])))
        
    chain.append(F.sigmoid(getattr(model, 'gen_%i' % (n_layers_gen))(chain[-1])))
        
    output = chain[-1]
    
    return output    
```

```python
def forward(x_data, train=True):
    inpt = chainer.Variable(x_data)
    
    chain = [inpt]
    for i in range(n_layers_recog):
        chain.append(F.relu(getattr(model, 'recog_%i' % i)(chain[-1])))
        
    code = getattr(model, 'recog_%i' % n_layers_recog)(chain[-1])
    mean, std = F.split_axis(code, 2, 1)
    std = (std ** 2 + 1e-4) ** .5
    
    sn_samples = StandardNormalSamples()(mean)
    
    sample = sn_samples * std + mean
    chain += [code, sample]
        
    for i in range(n_layers_gen):
        chain.append(F.relu(getattr(model, 'gen_%i' % i)(chain[-1])))
        
    chain.append(F.sigmoid(getattr(model, 'gen_%i' % (n_layers_gen))(chain[-1])))
        
    output = chain[-1]
    # Chainer divides by the whole size for MSE, which is why we have 
    # to multiply with the data dimension.
    rec_loss = F.mean_squared_error(output, inpt) * x_data.shape[1]
    
    kl_loss = F.basic_math.Log()(1 / std + 1e-4) + (std ** 2 + mean ** 2) / 2 - .5
    
    loss = F.Sum()(rec_loss) + F.Sum()(kl_loss)
    
    return loss, output    
```

```python
optimizer = optimizers.Adam()
optimizer.setup(model.collect_parameters())
```

```python
for epoch in six.moves.range(1, n_epochs + 1):
    print('epoch', epoch)

    optimizer.zero_grads()
    batches = iter_minibatches([x_train], batch_size, [0], 1)
    
    for batch, in batches:
        optimizer.zero_grads()
        loss, output = forward(batch)
        loss.backward()
        optimizer.update()
        
    loss, _ = forward(x_train)
    print loss.data
```
